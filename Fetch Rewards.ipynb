Use pandas to clean the data

```python
import json
import pandas as pd
from ast import literal_eval

from datetime import datetime
```

Open receipts, brands and users table.

```python
receipts = pd.read_json('receipts.json',lines = True)
brands = pd.read_json('brands.json',lines = True)
users = pd.read_json('users.json',lines = True)
```

### Receipts table:

```python
receipts.head()
```

Column 'rewardsReceiptItemList' contains lists of variables. We have to split each of them and then join to the original receipts table.

```python
receipts = receipts.explode('rewardsReceiptItemList')
receipts.reset_index(inplace = True)
```

```python
receipts = receipts.fillna({'rewardsReceiptItemList':'{}'})

receipts['rewardsReceiptItemList'] = receipts['rewardsReceiptItemList'].apply(lambda x:str(x))
receipts['rewardsReceiptItemList'] = receipts['rewardsReceiptItemList'].apply(literal_eval)
#raises an exception if the input isn't a valid Python datatype, so the code won't be executed if it's not.
```

```python
#Normalize semi-structured JSON data into a flat table.
receipts_norm = pd.json_normalize(receipts['rewardsReceiptItemList'],errors = 'ignore',record_prefix = 'rewardsReceiptItemList')\
.add_prefix('rewardsReceiptItemList.')
```

```python
receipts_df = pd.merge(receipts, receipts_norm, left_index = True, right_index = True, how = 'outer')
```

```python
receipts_df
```

After the split for 'rewardsReceiptItemList', since there are dictionaries existed, we have to reformat.

```python
set().union(*(d.keys() for d in receipts_df['_id']))
set().union(*(d.keys() for d in receipts_df['createDate']))
set().union(*(d.keys() for d in receipts_df['dateScanned']))
set().union(*(d.keys() for d in receipts_df.dropna(subset=['finishedDate'])['finishedDate']))
set().union(*(d.keys() for d in receipts_df.dropna(subset=['modifyDate'])['modifyDate']))
set().union(*(d.keys() for d in receipts_df.dropna(subset=['pointsAwardedDate'])['pointsAwardedDate']))
set().union(*(d.keys() for d in receipts_df.dropna(subset=['purchaseDate'])['purchaseDate']))
```

Get rid of unwanted characters and set adjust time in a standard way.

```python
def date_converter(x):
    try:
        return(datetime.utcfromtimestamp(int(x['$date'])/1000).strftime('%Y-%m-%d %H:%M:%S'))
    except TypeError:
        return(None)
```

```python
receipts_df['_id'] = receipts_df['_id'].apply(lambda x: x['$oid'])
receipts_df['createDate'] = receipts_df['createDate'].apply(lambda x: date_converter(x))
receipts_df['dateScanned'] = receipts_df['dateScanned'].apply(lambda x: date_converter(x))
receipts_df['finishedDate'] = receipts_df['finishedDate'].apply(lambda x: date_converter(x))
receipts_df['modifyDate'] = receipts_df['modifyDate'].apply(lambda x: date_converter(x))
receipts_df['pointsAwardedDate'] = receipts_df['pointsAwardedDate'].apply(lambda x: date_converter(x))
receipts_df['purchaseDate'] = receipts_df['purchaseDate'].apply(lambda x: date_converter(x))
```

```python
receipts_df.head()
```
